```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import string
%matplotlib inline
```

```python
# check for missing values
df.isna().sum()
```

```python
# Convert actual_price column to string (in case it's not)
df['actual_price'] = df['actual_price'].astype(str)

# Remove non-numeric characters like ‚Çπ, commas, and dashes
df['actual_price'] = df['actual_price'].str.replace(r'[^\d.]', '', regex=True)

# Convert the cleaned strings to numeric values; invalid parsing will be set as NaN
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Fill missing values (NaNs) with the mean of the column
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].mean())

# Check how many missing values remain (should be 0)
print(df['actual_price'].isna().sum())
```

```python

# If the column is stored as strings, we try to extract the first float number in each cell
import re

def extract_first_rating(x):
    try:
        # Find the first valid float number in the string
        match = re.search(r'\d+(\.\d+)?', str(x))
        return float(match.group()) if match else None
    except:
        return None

# Apply this function to the 'rating' column
df['rating'] = df['rating'].apply(extract_first_rating)

# Then fill missing values with the mean
df['rating'] = df['rating'].fillna(df['rating'].mean())

# Confirm there are no more NaNs
print(df['rating'].isna().sum())
```

```python

# Convert the 'rating_count' column to numeric values.
# Any entries that can‚Äôt be parsed as numbers will become NaN.
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Replace all NaN values in 'rating_count' with the column‚Äôs mean.
# This fills in missing entries while preserving the overall distribution.
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mean())
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].mean())
df['rating'] = df['rating'].fillna(df['rating'].mean())
# Verify that there are no remaining NaN values in 'rating_count'.
# This should output 0 if all missing values have been filled.
df["rating_count"].isna().sum()
df["actual_price"].isna().sum()
```

```python
# List of columns that need to be cleaned and converted to numeric values
columns_to_clean = ['discounted_price', 'actual_price', 'discount_percentage','rating',  'rating_count']

# Remove unwanted characters like ‚Çπ, %, and commas
for col in columns_to_clean:
    df[col] = df[col].astype(str).str.replace('‚Çπ', '', regex=False)    # Remove currency symbol
    df[col] = df[col].str.replace(',', '', regex=False)                # Remove commas
    df[col] = df[col].str.replace('%', '', regex=False)                # Remove percentage symbol
    df[col] = pd.to_numeric(df[col], errors='coerce')                  # Convert to numeric, invalid values become NaN

# Fill missing values in rating_count with 0
df['rating_count'] = df['rating_count'].fillna(0)

# Quick check of dtypes to confirm
print(df[['discounted_price','actual_price','discount_percentage','rating','rating_count']].dtypes)
```

```python
#  Count total duplicate rows across all columns
total_dupes = df.duplicated().sum()
print(f"Total duplicate rows (all columns): {total_dupes}")

#  Inspect the first few duplicate rows to see what‚Äôs repeating
dupe_rows = df[df.duplicated(keep=False)]   # keep=False flags all copies
print(dupe_rows.head())
```

```python
#  Define the grouping key (product)
key_cols = ['product_id']

#  For each product_id, find the index of the row with max rating_count
best_idx = df.groupby(key_cols)['rating_count'].idxmax()

#  Create a new DataFrame with only those ‚Äúbest‚Äù rows
df_best = df.loc[best_idx].reset_index(drop=True)

#  Inspect the result
print(df_best.head())
```

```python
# Quick look at the new table
print(df_best.shape)     # see how many products you have
print(df_best.head())    # glimpse of columns

# Sort by rating_count descending to see your highest-voted products
df_best_sorted = df_best.sort_values(by='rating_count', ascending=False)
print(df_best_sorted[['product_id','rating_count','rating']].head(10))
```

```python
# Make sure price columns are numeric (we cleaned them already)
summary = df_best_sorted[['product_id','rating_count','rating']].merge(
    df_best_sorted[['product_id','discounted_price','actual_price','discount_percentage']],
    on='product_id',
    how='left'
)
print(summary.head(10))
```

        Summerize Data using Descriptive Statistics
</p>


```python
from IPython.display import display

#  Ensure 'rating' is numeric in case it's still an object
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

#  Define which numeric columns to summarize
numeric_cols = [
    'discounted_price',
    'actual_price',
    'discount_percentage',
    'rating',

]

#  Compute descriptive statistics
stats = df[numeric_cols].agg(['mean', 'median', 'min', 'max', 'std'])

#  Add a 'range' row
stats.loc['range'] = stats.loc['max'] - stats.loc['min']

#  Transpose for a clear, feature-by-statistic layout
stats = stats.T
stats.columns.name = 'Statistic'
stats.index.name = 'Feature'


# Round for readability
stats = stats.round(3)

#  Display as a neat table in Jupyter
display(stats)
```
