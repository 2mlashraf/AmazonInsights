```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import string
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Basmala\OneDrive\Desktop\Amason\Data\amazon.csv")
df.head()
```

```python
df['category'].unique()
```

```python
# clean category column 
df['Cleaned_Category'] = df['category'].apply(lambda x: '|'.join(x.split('|')[:2]))
df['category'] = df['Cleaned_Category']

df.drop('Cleaned_Category', axis=1, inplace=True)
df['category'].unique()
```

```python
df['user_name'].unique()
```

```python
# Clean user_name column to keep only the first name
df['Cleaned_name'] = df['user_name'].apply(lambda x: x.split(',')[0])  # Keep only the first name
df['user_name'] = df['Cleaned_name']  # Replace the original user_name with cleaned version

df.drop('Cleaned_name', axis=1, inplace=True)  # Drop the temporary cleaned column
df['user_name'].unique()  # Check the unique values after cleaning
```

```python
df.info()
```

```python
pd.DataFrame({
    'non-null': df.count(),
    'unique'  : df.nunique(),
    'top'     : df.mode().iloc[0],
    'freq'    : df.apply(lambda x: x.value_counts().iloc[0])
})
```

```python
import re
columns_to_clean = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

for col in columns_to_clean:
    df[col] = df[col].astype(str) 
    df[col] = df[col].str.replace(r'[₹$,%,]', '', regex=True) 
    df[col] = pd.to_numeric(df[col], errors='coerce')  

def extract_first_rating(x):
    try:
        match = re.search(r'\d+(\.\d+)?', str(x))
        return float(match.group()) if match else None
    except:
        return None

df['rating'] = df['rating'].apply(extract_first_rating)


print(df[columns_to_clean].isna().sum())
```

```python
df = df.dropna(subset=['actual_price', 'rating', 'rating_count'])
```

```python
# check for missing values
df.isna().sum()
```

```python
df.head()
```

```python
df.describe()
```

```python
#  Count total duplicate rows across all columns
total_dupes = df.duplicated().sum()
print(f"Total duplicate rows (all columns): {total_dupes}")

#  Inspect the first few duplicate rows to see what’s repeating
dupe_rows = df[df.duplicated(keep=False)]   # keep=False flags all copies
print(dupe_rows.head())
```

```python
#  Define the grouping key (product)
key_cols = ['product_id']

#  For each product_id, find the index of the row with max rating_count
best_idx = df.groupby(key_cols)['rating_count'].idxmax()

#  Create a new DataFrame with only those “best” rows
df_best = df.loc[best_idx].reset_index(drop=True)

#  Inspect the result
df_best.head()
```

```python
# Quick look at the new table
print(df_best.shape)     # see how many products you have
print(df_best.head())    # glimpse of columns

# Sort by rating_count descending to see your highest-voted products
df_best_sorted = df_best.sort_values(by='rating_count', ascending=False)
df_best_sorted[['product_id','rating_count','rating']].head(10)
```

```python
# Make sure price columns are numeric (we cleaned them already)
summary = df_best_sorted[['product_name','rating_count','rating']].merge(
    df_best_sorted[['product_name','discounted_price','actual_price','discount_percentage']],
    on='product_name',
    how='left'
)
summary.head(10)
```

<p style="font-family:Arial; font-size:30px; color:white; background-color:black; padding:10px; border-radius:10px;">
        Summerize Data using Descriptive Statistics
</p>


```python
from IPython.display import display

#  Ensure 'rating' is numeric in case it's still an object
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

#  Define which numeric columns to summarize
numeric_cols = [
    'discounted_price',
    'actual_price',
    'discount_percentage',
    'rating',

]

#  Compute descriptive statistics
stats = df[numeric_cols].agg(['mean', 'median', 'min', 'max', 'std'])

#  Add a 'range' row
stats.loc['range'] = stats.loc['max'] - stats.loc['min']

#  Transpose for a clear, feature-by-statistic layout
stats = stats.T
stats.columns.name = 'Statistic'
stats.index.name = 'Feature'


# Round for readability
stats = stats.round(3)

#  Display as a neat table in Jupyter
display(stats)
```

#### Difference Between Custom Statistics Code and `describe()`

| Aspect | Custom Statistics Code | `describe()` |
|:------|:------------------------|:-------------|
| **Selected Columns** | Manually select specific columns to summarize. | Summarizes **all numeric columns** automatically. |
| **Calculated Metrics** | Calculates `mean`, `median`, `min`, `max`, `std`, and **range**. | Calculates `count`, `mean`, `std`, `min`, `25%`, `50% (median)`, `75%`, `max`. |
| **Range Calculation** | Explicitly added (`max - min`). | Not calculated by default. |
| **Data Presentation** | After transposing, features become rows and statistics become columns. | Features are columns, statistics are rows. |
| **Rounding** | Results are rounded (e.g., to 3 decimal places) for clarity. | No rounding by default. |
| **Flexibility** | Fully customizable (metrics, formatting, etc.). | Standard output, less customizable. |

---

> **Summary:**  
> - Use the **custom code** for greater flexibility and when specific statistics or formatting are needed.  
> - Use **`describe()`** for a quick, general overview of the data.
