## **Summary of Findings**

### **1. Price Analysis**
- The dataset contains products with a wide range of prices:
  - Example: **SKYWALL 81.28 cm (32 inches) HD Ready Smart LED TV** has an actual price of ₹19,125.
  - Example: **Duracell USB Lightning Apple Certified Cable** is priced at ₹1,799.
  
- Discounted prices also show significant variations:
  - Example: **Zoul USB C 60W Fast Charging Cable** is priced at ₹389 (discounted from ₹1,099), showing a **65%** discount.
  
- The price differences reflect various types of products in the dataset, with some offering heavy discounts while others remain at a higher price point.

### **2. Discounts**
- Discounts range from **46%** to **70%** off the actual price:
  - Example: **Duracell USB Cable** offers a **46%** discount.
  - Example: **Zoul USB C Cable** offers a **65%** discount.
  
- Discounts on products may influence the perceived value and attractiveness to customers. Products offering significant discounts (e.g., **Zoul USB C Cable**) show potential to drive more sales.

### **3. Ratings and Reviews**
- Ratings vary across products, from **3.4** to **4.5**:
  - Example: **Duracell USB Cable** has a high rating of **4.5**.
  - Example: **SKYWALL Smart LED TV** has a lower rating of **3.4**.
  
- **Rating count** plays a role in product popularity:
  - Example: **Duracell USB Cable** has **815 reviews**, while **SKYWALL TV** has **902 reviews**.
  - Higher ratings and review counts generally indicate higher customer satisfaction and product popularity.

### **4. Categories**
- Products span across multiple categories:
  - Examples include **Computers & Accessories**, **Electronics**, and **Home Theater**.
  
- The most frequent categories are **Accessories & Peripherals** and **Televisions**.

### **5. Outliers and Anomalies**
- Some products show extreme values for certain metrics, which are flagged as outliers:
  - **Example**: Products with high prices and low ratings (e.g., **SKYWALL TV**) could indicate that despite high cost, the product is not well-received by customers.
  - **Example**: **CROSSVOLT Compatible Dash Cable** has a low price but high rating, making it a valuable product for customers.
  
- These anomalies should be further explored for understanding customer expectations versus product performance.

### **6. Product Descriptions & Reviews**
- **Review titles and content** indicate customer satisfaction:
  - Example: **CROSSVOLT Dash Cable** receives feedback such as "No reasons to complain" and "Value for money."
  
- Some products also receive constructive criticism:
  - Example: **Zoul USB Cable** received comments like "Charging speeds could be better," indicating areas of improvement despite positive feedback.

### **7. Images and Links**
- **Product images** and **official product links** are included in the dataset, which are key in the product's marketing and decision-making process.
  - Example: Each product has an image link, e.g., for **CROSSVOLT**, a high-quality image is available.
  
- Links to the product pages help customers make informed decisions by leading them directly to official purchasing sites.

## **Key Insights**
- **Products with higher ratings** tend to have more reviews and are often better priced.
- **Discounted products** are attractive, but the **value for money** perception plays a significant role in customer satisfaction.
- **Outliers** such as **expensive products with low ratings** suggest that some high-cost products may not meet customer expectations despite their price.
